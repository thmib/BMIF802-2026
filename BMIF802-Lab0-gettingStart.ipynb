Below, we write a function that gives the tax rate based on age and annual income (as inputs)

```python
# A function that calculates the appropriate tax rate
def get_tax_rate(age, income):    
    # now we compute the tax rate
    if age<65:
        if income<=41000:
            tax_rate=0.15
        else:
            tax_rate= 0.22
    else: # age>=65
        if income<=47000:
            tax_rate=0.15
        else:
            tax_rate=0.22

    return tax_rate
    

def main():
    # take inputs from the user
    age = int(input("Please enter age:"))
    income = int(input("Please enter annual income:"))    
    result = get_tax_rate(age, income)
    
    # print the result
    print("The tax rate is",result)

    
main()
```
